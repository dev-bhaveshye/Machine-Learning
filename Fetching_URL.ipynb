```python
import requests
responce =requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d77ec00fcc5712f582efcd84242f8146&language=en-US&page=1')
```

```python
DF=pd.DataFrame()
DF
```

```python
for i in range(1, 490):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=d77ec00fcc5712f582efcd84242f8146&language=en-US&page={i}')


    if response.status_code == 200:
        temp_df = pd.DataFrame(response.json()['results'])[['title', 'vote_average', 'id', 'overview', 'release_date', 'popularity', 'vote_count']]
        DF = pd.concat([DF, temp_df], ignore_index=True)
    else:
        print(f"Request failed for page {i} with status code: {response.status_code}")
        print(response.json())
```
