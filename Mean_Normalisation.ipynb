**Mean_Normalisation( for centered data )**

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Salary_dataset (1).csv',header=0,usecols=['YearsExperience','Salary'])
df
```

```python
import numpy as np
```

```python
mean = df['Salary'].mean()
mean
```

```python
max=df['Salary'].max()
max
```

```python
min=df['Salary'].min()
min
```

```python
x=(df['Salary']-mean)/(max-min)
x
```

```python
import seaborn as sns
sns.scatterplot(x)
```

```python
sns.distplot(x,hist=False)
```

**Robust Normalization ( for outliers )**

```python
!pip install scikit-learn
```

```python
from sklearn.preprocessing import RobustScaler
robust=RobustScaler()
robust.fit(df[['Salary']])
robust.transform(df[['Salary']])
```

```python
robust_scaled=pd.DataFrame(robust.transform(df[['Salary']]),columns=['Salary'])
robust_scaled
```

```python
sns.distplot(robust_scaled)
```
