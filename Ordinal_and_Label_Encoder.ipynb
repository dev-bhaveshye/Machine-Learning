```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
oe_day=oe.fit_transform(df[['day']])
oe_day
```

```python
oe_day=pd.DataFrame(oe_day,columns=['day'])
```
